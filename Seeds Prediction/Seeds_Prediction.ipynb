# Dataset

https://archive.ics.uci.edu/ml/machine-learning-databases/00236/

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['area', 'perimeter', 'compactness', 'length', 'width', 'asymmetry', 'groove', 'class']

df = pd.read_csv('seeds_dataset.txt', names=cols, sep="\s+")
df.head()
```

```python
# Consider 'class' doesn't exist in the given dataset, as this analysis will be for unsupervised learning

for i in range(len(cols)-1):
  for j in range(i+1, len(cols)-1):
    x_label = cols[i]
    y_label = cols[j]
    sns.scatterplot(x=x_label, y=y_label, data=df, hue='class')
    plt.show()
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
x = 'perimeter'
y = 'asymmetry'
X = df[[x, y]].values
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
clusters = kmeans.labels_
clusters
```

```python
df['class'].values
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1,1))), columns = [x,y,'class'])
```

```python
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df)
plt.plot()
```

```python
# Higher dimensions

X = df[cols[:-1]].values
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
cluster_df = pd.DataFrame(np.hstack((X, kmeans.labels_.reshape(-1,1))), columns = df.columns)
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df)
plt.plot()
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
transformed_X = pca.fit_transform(X)
```

```python
X.shape
```

```python
transformed_X.shape
```

```python
transformed_X[:5]
```

```python
plt.scatter(transformed_X[:, 0], transformed_X[:,1])
plt.show()
```

```python
kmeans_pca_df = pd.DataFrame(np.hstack((transformed_X, kmeans.labels_.reshape(-1,1))), columns=['pca1', 'pca2', 'class'])
```

```python
true_pca_df = pd.DataFrame(np.hstack((transformed_X, df['class'].values.reshape(-1,1))), columns=['pca1', 'pca2', 'class'])
```

```python
sns.scatterplot(x='pca1', y='pca2', hue='class', data=kmeans_pca_df)
plt.plot()
```

```python
sns.scatterplot(x='pca1', y='pca2', hue='class', data=true_pca_df)
plt.plot()
```
