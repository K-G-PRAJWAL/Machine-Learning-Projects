## Prediction using Supervised Machine Learning

```python
# Import packages
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Read dataset
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head(5)
```

Let's plot our data points on 2-D graph to eyeball our dataset and see if we can manually find any relationship between the data. We can create the plot with the following script:

```python
# Plotting the graph
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

There is a positive linear relation between the number of hours studied and percentage of score.

```python
# Divide dataset into features = X and label = Y
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
```

```python
# Split dataset into training and testing
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python
# Use linear regression from sci-kit learn library to train the model
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
```

```python
# Fit the model on training data
regressor.fit(X_train, y_train) 
```

```python
# Plot the regression line
line = regressor.coef_ * X + regressor.intercept_
```

```python
# Plot for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

```python
print(X_test)
```

```python
# Make predictions on testing data
y_pred = regressor.predict(X_test)
```

```python
# Compare Actual vs Predicted label
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# Evaluate model using Mean Absolute Error
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
