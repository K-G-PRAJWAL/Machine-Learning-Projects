# Movie Recommender System (<font color="green">Movie Similarities</font>)

This project will be using the MovieLens dataset provided by grouplens.org and groups movies 

* Use pandas to load the rows of the u.data and u.item files and merge them to work with movie names and not their ID's.

```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

* Use the pivot_table function on the Dataframe to construct a user/movie rating matrix.

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

* Extract a series of users who rated Star Wars.

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
```

* Use the pandas corrwith function to compute pairwise correlation of Star Wars vector of user rating with every other movie.
* Drop any results that have no data.
* Construct a new Dataframe of movies and their correlation score to Star Wars.

```python
similarMovies = movieRatings.corrwith(starWarsRatings) # pairwise correlation of Star Wars vector of user rating with every other movie
similarMovies = similarMovies.dropna() # Drop any results that have no data
df = pd.DataFrame(similarMovies) # Construct a new Dataframe of movies and their correlation score to Star Wars
df.head(10)
```

* Sort the results by similarity score.

```python
similarMovies.sort_values(ascending=False)
```

* Yes, the results does not make sense.
* This is because our results are being messed up by movies that have only been viewed by a handful of people who also happened
to like Star Wars.
* Lets get rid of the movies that were only watched by a few people that are producing wrong results.

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

* Get rid of movies rated by fewer than 100 people.

```python
popularMovies = movieStats['rating']['size'] >= 100 # Ignore movies rated by less than 100 people
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

* Join this data with original set of similar movies to Star Wars.

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
```

```python
df.head()
```

* Sort them by similarity sore.

```python
df.sort_values(['similarity'], ascending=False)[:15]
```

Therefore, we get all the Star Wars related movie as the top 3 entries.
