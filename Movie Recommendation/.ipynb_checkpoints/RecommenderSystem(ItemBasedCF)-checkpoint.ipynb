# Movie Recommender System(<font color="green">Item-Based Collaborative Filtering</font>)

The aim of this project will be to provide recommendations to the user based of the activities of all the other users.

* Import the MovieLens 100K data set into a pandas DataFrame.

```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)

ratings.head()
```

* Pivot the table to create a matrix of users and the movies they rated.

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.head()
```

* Use the pandas corr() method to compute a correlation score for every column pair in the matrix.

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

* Use min_periods argument to remove results of movies rated by users lesser than 100 in number.

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

* Produce some movie recommendations for user ID 0(test case).

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

* For each movie rated, retreieve list of similar movies for the correlation matrix.

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print ("Adding sims for " + myRatings.index[i] + "...")
    # Retrieve similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at our results so far:
print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

* Remove duplicate entries.

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

* Filter movies that have been already rated.

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
